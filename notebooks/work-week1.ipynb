## 3. 사용할 데이터셋 결정하기
* 원래는 Yahoo Finance API를 사용하려 했으나, 2주차 수업 후 Github Archive데이터로 변경

```python
import os
import subprocess

# 다운로드할 폴더 설정
folder = './gh_archive/'
folder = '../data/'
os.makedirs(folder, exist_ok=True)

# 날짜와 시간을 반복하여 파일 다운로드
for day in range(1, 2):
    for hour in range(0, 24):
        url = f'https://data.gharchive.org/2015-01-{day:02d}-{hour:02d}.json.gz'
        #print(url)
        subprocess.run(['wget', '-P', folder, url])
```
